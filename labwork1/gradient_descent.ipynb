```python
def f(x):
    return x**2
```

```python
def f_(x):
    return 2*x
```

```python
def gradient_descent(init, lr, threshold):
    x = init
    i = 0
    diff = 999
    
    while diff > threshold: 
        i = i+1
        x_a = x
        x = x - lr*f_(x)
        diff = abs(x - x_a) 
        print(f"Time: {i}, x: {x:.5f}, f(x): {f(x):.5f}")
```

```python
gradient_descent(10, 0.9, 0.001)
```
